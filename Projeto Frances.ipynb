```python
import pandas as pd
import matplotlib.pyplot as plt

#Para fazer a alteração do planilha, basta colocar a nova planilha na mesma pasta e trocar o nome do arquivo selecionado para o arquivo desejado ex.:('nome_do_arquivo.xlsx')
data = pd.read_excel('frances_agosto_08.xlsx')

data.info()

contagem = data['MUSICA'].value_counts()

top_20_frances = data['MUSICA'].value_counts().head(20)

top_20_frances = top_20_frances.reset_index()

top_20_frances.columns = ['ARTISTA - MUSICA', 'MAIS TOCADA']
    
plt.figure(figsize=(10, 8))
top_20_frances.plot(kind='bar', color='skyblue')
plt.title('Top 20 mais tocadas da Rádio Francês FM em Agosto "MUSICA"')
plt.xlabel('ARTISTA')
plt.ylabel('MAIS TOCADA')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
top_20_frances
```

```python
# Caro Radialista, esse código apresentará as músicas mais tocadas no dia selecionado
#para fazer a alteração da data, basta so trocar a informação de data que consta abaixo ex.: dia_tocado = '2024-08-20'
dia_tocado = '2024-08-20'
tocada_por_dia = data['DATA'] == dia_tocado

#Abaixo será apresentada todas as músicas que foram tocadas no dia 20 de Agosto
data[tocada_por_dia]
```

```python
#apresenta as tops 20 músicas que foram mais tocadas no dia 20 de Agosoto;
top_dia =data[tocada_por_dia].value_counts('MUSICA').head(20)

top_dia = top_dia.reset_index()
top_dia.columns = ['Artista', 'Total dia 20-08']

top_dia
```

```python
#Para pesquisar quantas vezes a música foi tocada no mês, basta selecionar o nome do arquivo da música como consta abaixo
musica_mes = data['MUSICA'] == 'GUSTAVO MIOTO - VOLTANDO PRA MIM'
data[musica_mes].head(5)
```

```python
por_dia =data[musica_mes].value_counts('DATA').head(20)
por_dia.columns= ['DATA', 'VEZES']
plt.figure(figsize=(10, 8))
por_dia.plot(kind='bar', color='orange')
plt.title('Acompanhamento Diário Gustavo Mioto - Voltando pra mim, Rádio Francês FM em Agosto "DATA"')
plt.xlabel('DIA')
plt.ylabel('VEZES')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Exibindo o gráfico

plt.show()
```
